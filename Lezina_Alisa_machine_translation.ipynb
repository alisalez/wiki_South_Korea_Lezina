#### 1. Выбранная пара языков: коррейский - русский.

Строй этих языков сильно различается. Корейский - агглютинативный язык с (в большинстве случаев) фиксированным порядком слов SOV, русский - флективный язык с (в большинстве случаев) свободным порядком слов (базовым считается порядок SVO).

#### 2. Пре-процессинг

__Этап 1: токенизация.__
- Для русского - разделение текста на слова и знаки препинания.
- Для корейского - выделение лексем (словосочетаний), поскольку слова могут быть "слиты".

__Этап 2: лемматизация и морфологический анализ__
- Русский язык: определение основы слова и его грамматических характеристик (например, род, число и падеж для существительного).
- Корейский язык: разбор окончаний и частиц (анализ агглютинативной структуры).

__Этап 3: синтаксический разбор__
- Построение деревьев зависимостей, описывающих синтаксис предложений в обоих языках.

__Этап 4: устранение неоднозначностей__
- Выбор корректного перевода для многозначных слов и словосочетаний.

*__Библиотеки для пре-процессинга__*
- Для русского языка:  
  - pymorphy2 (морфологический анализ)  
  - nltk или spacy (токенизация и синтаксический разбор)
- Для корейского языка:  
  - konlpy (морфологический анализ и токенизация)  
  - KoNLPy с морфоанализаторами, такими как Kkma, Komoran, или Mecab.

#### 3. Примеры правил

__Правило 1: Порядок слов__
- Корейский: SOV (Я кофе пью).  
- Русский: SVO (Я пью кофе).

__Пример:__
S (корейский): 저는 커피를 마셔요

- 저는 - я с выделительной частицей 은/는
- 커피를 - кофе с частицей 을/를 винительного падежа
- 마셔요 - пью (настоящее время, неофициально вежливый стиль)

__Правило:__
1. Определить подлежащее (S) — 저는 → "Я".  
2. Определить дополнение (O) — 커피를 → "яблоко".
3. Определить сказуемое (S) — 마셔요 → "пью".
4. Переставить на порядок SVO: Я пью кофе.

__Правило 2: Перевод частиц__  
- Частицы в корейском (-을/를, -이/가, -에 и др. ) обозначают падежи. В русском эти значения выражаются через окончания или предлог + окончание.  

__Пример:__  
Корейский: 책을 읽다 (책 - книга, 을 - винительный падеж, 읽다 - читать).  
Перевод:  
1. Распознать частицу 을 как маркер винительного падежа.  
2. Перевести слово 책 как "книга" и преобразовать в винительный падеж: книгу.  
3. Результат: Читать книгу.

__Правило 3: Обработка времени глаголов__
- В корейском языке время выражается через суффиксы (например, -는다 для настоящего времени), а в русском — через спряжение.  

Пример:  
Корейский: 나는 학교에 갔다  
(나는 - я, 학교에 - в школу, 갔다 - ходил).  

1. Определить форму 갔다 как прошедшее время (окончание __-았다__).  
2. Перевести основу глагола 가- как "ходить".  
3. Спрягать глагол: "Я ходил в школу".
