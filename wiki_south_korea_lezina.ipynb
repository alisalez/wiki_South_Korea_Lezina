```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/South_Korea'
r_wiki = requests.get(url=url)
```

```python
r_wiki
```

```python
r_wiki.text
```

```python
soup_wiki = BeautifulSoup(r_wiki.text, 'html.parser')
soup_wiki
```

```python
pars_wiki = soup_wiki.find_all('p')
pars_wiki
```

```python
texts_wiki = [text.get_text() for text in pars_wiki]
texts_wiki
```

```python
texts_wiki = '\n'.join([x for x in texts_wiki])
```

```python
data = texts_wiki
with open('text.txt','w') as f:
  f.write(data)
```
