SMT (Statistical Machine Translation) -  метод машинного перевода, который основывается на статистике. Цель состоит в том, чтобы перевести предложение с одного языка на другой __наиболее правдоподобным образом__.

###__Компоненты SMT__
Формальное определение SMT (математика):

---

#### 1. __В чем цель?__
Мы хотим перевести предложение с одного языка \( X \) (например, с русского) на другой язык \( Y \) (например, английский) __максимально правдоподобно__.  

P.s.  "Правдоподобно" означает, что результат должен быть похож на то, как люди говорят или пишут на языке \( Y \).  

Для этого мы ищем такое предложение \( y \) на языке \( Y \), которое __максимизирует вероятность__ того, что оно правильно переведено: __argmax P(y|x)__.

(где P(y|x) — это вероятность того, что \( y \) является правильным переводом для \( x \)).

---

#### 2. __Поиск argmax P(y|x)__

1. Вычислить допустимые варианты предложений на языке Y путем создания __языковой модели P(y)__.  
2. Вычислить допустимые пары X - Y путем __сопоставления статистических совпадений__ выравненных фраз в параллельном корпусе P(x|y).
3. __Максимизировать результат__ вычисления P(y) P(x|y), т.е. argmax P(y) P(x|y)

Компоненты модели SMT (код):

---

1. __N-граммная языковая модель \( P(y) \)__:
- моделируем __вероятностное распределение конструкций__ на уровне слов или фраз в языке Y  
__Проще:__ модель говорит, насколько предложение на языке \( Y \) похоже на настоящее, "естественное" предложение.  

2. __Модель перевода aka t-model \( P(x|y) \)__:  
- Cобираем статистику соотвествий фраз в паралелльном корпусе, ищем переводческие соответствия X - Y и моделируем их с помощью теории вероятности.
__Проще:__ модель говорит, насколько хорошо предложение \( y \) переводит \( x \).

3. Допускаем, что любое предложение языка Y может быть "искаженной" версией некой фразы на языке X.

4. Ищем наиболее правдоподобные соответствия X - Y.

5. __Декодер__: ищем наиболее грамматичные и лексически правдоподобные результаты, отбираем среди гипотез один результат
